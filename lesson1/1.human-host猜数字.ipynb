这里做一个简单的human与host猜数字的对话，host生成10以内数字human猜中对话结束。

生成句子的测试用例见preparatory/3.Random Chinese Sentence Generator.ipynb

```python
#输入字符串，提取规则返回字典。这里传入的参数还有输入字符串中划分statement和expression、expression之间分隔符。
def map(simple_grammar,stmt_split='=', or_split='|'):
    rules = dict() # key is the @statement, value is @expression
    for line in simple_grammar.split('\n'):
        if not line: 
            continue   # skip the empty line
        #print(line)
        stmt, expr = line.split(stmt_split)
        expr=expr.replace('null',' ')    #不替换成空，为了下一行split可以分开，空格之后可以处理掉。其实这里null后面本来就有空格
        #print(expr)
        rules[stmt.strip()] = expr.split(or_split)
    return rules
```

```python
import random
def generate(rules,target='sentence',trace=0):#trace代表是否追踪生成过程
    global i,j
    if target in rules:                       #target是statement
        candidates = rules[target]            
        candidate = random.choice(candidates) 
        if trace==1:
            print('调用函数第%d次，target=%s，candidate=%s'%(i,target,candidate))
        i+=1        
        return ''.join(generate(rules, target=c.strip(),trace=trace) for c in candidate.split())  #c.strip()把每个expression中的空格去除
    else:                                 #target不是statement，为递归边界直接返回target
        if trace==1:
            print('调用函数第%d次，target=%s，达到递归边界得到生成的第%d个词---------------------我是分割线----------------------'%(i,target,j))
        i+=1
        j+=1
        return target
```

```python
def dialogue_n(simple_grammar,target='sentence',stmt_split='=', or_split='|',trace=0):
    rules=map(simple_grammar,stmt_split=stmt_split, or_split=or_split)              #提取规则
    global i,j                             
    i=1
    j=1
    print('%s:'%simple_grammar.split('=')[0].replace('\n','').strip())
    o=generate(rules, target=target,trace=trace)   #递归生成句子
    print(o)
    return o
```

```python
human = """
human = 自己 寻找 活动 结尾
猜数 = 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 | 9 
自己 = 我 | 俺 | 我们 
寻找 = 想看看 | 想找找 | 想找点
活动 = 乐子，那就猜数字 | 玩的，那就猜数字
结尾 = 吧。 | 好了。 |这样。
"""
```

```python
host = """
host = 寒暄 报数 询问 业务相关 结尾 
生成数字 = 数字已经生成，试试吧。
重复 = 这个数字您之前已经猜过啦, 错
猜对 = 祝贺 对
猜错 = 恭维 错
祝贺 = 太棒了！ | 你好厉害， | 你太强了，
对 = 你猜对了！ | 猜对啦。 | 命中答案。
恭维 = 好可惜， | 还差一点, | 很接近了,
错 = 再猜一次吧。|猜错啦。| 下次就能猜中。 
报数 = 我是 数字 号 ,
数字 = 单个数字 | 数字 单个数字 
单个数字 = 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 | 9 
寒暄 = 称谓 打招呼 | 打招呼
称谓 = 人称 ,
人称 = 先生 | 女士 | 小朋友
打招呼 = 你好 | 您好 
询问 = 请问你要 | 您需要
业务相关 = 玩玩 具体业务
玩玩 = 耍一耍 | 玩一玩
具体业务 = 猜数字
结尾 = 吗？"""
```

```python
dialogue_n(host,target='host')
dialogue_n(human,target='human')
n=random.randint(1,10)
#这里可以日后拓展，提取出human的意图再进行下一步工作
dialogue_n(host,target='生成数字')
a=[]                                  #记录human猜过的数字
while True :
    temp=dialogue_n(human,target='猜数')[-1]
    if temp==str(n):
        dialogue_n(host,target='猜对')
        break
    else: 
        if temp in a:
            dialogue_n(host,target='重复')
        else :
            dialogue_n(host,target='猜错')
            a.append(temp)
        
```
